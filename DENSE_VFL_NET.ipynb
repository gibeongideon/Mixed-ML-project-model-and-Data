```python
#MODEL

from keras import backend as K
from __future__ import division, print_function 
from keras.callbacks import ModelCheckpoint 
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation,Flatten
from keras.optimizers import SGD,Adam,RMSprop
from keras.utils import np_utils
import numpy as np
import pylab as plt
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python

#np.random.seed(1671) # for reproducibility
# network and training
NB_EPOCH = 2
BATCH_SIZE = 65
MODEL_DIR = "/home/gsociety/Desktop/The Model/STORAGE/MODELS/DENSE"
VERBOSE = 1
NB_CLASSES = 1 # number of outputs = number of digits
OPTIMIZER =Adam() # optimizer, explained later in this chapter
N_HIDDEN = 112
N_HIDDEN1 =300
#INITIAL_EPOCH = 0
VALIDATION_SPLIT=0.25 # how much TRAIN is reserved for VALIDATION
DROPOUT = 0.05
# data: shuffled and split between train and test sets
#(X_train, y_train), (X_test, y_test) = mnist.load_data()
(x_train, y_train), (x_test, y_test)=vfl_data(t,10)

#X_train is 60000 rows of 28x28 values --> reshaped in 60000 x 784
RESHAPED = 21
#
x_train
```

```python
#
X_train = x_train.reshape(x_train.shape[0], RESHAPED)
X_test = x_test.reshape(x_test.shape[0], RESHAPED)
X_train = x_train.astype('float32')
X_test = x_test.astype('float32')
# normalize
X_train /= 50
X_test /= 50


# convert class vectors to binary class matrices..
#Y_train = np_utils.to_categorical(y_train, NB_CLASSES)
#Y_test = np_utils.to_categorical(y_test, NB_CLASSES)

Y_train = y_train
Y_test = y_test
#X_train
```

```python
#Hyper-parameteres optimal grid search


para = []
for n in range(11,100,9):
    for m in range(10,300,50):
        for r in range(100,700,100):
            para.append((n,m,r))

```

```python
#from time import sleep

#for n in para:
#    BATCH_SIZE = n[0]
 #   N_HIDDEN = n[1]
 #   N_HIDDEN1 =n[2]
   # sleep(10)
NB_EPOCH = 25

# M_HIDDEN hidden layers 10 outputs
model = Sequential()
model.add(Dense(30, input_shape=(RESHAPED,)))
model.add(Activation('relu'))
#model.add(Dropout(DROPOUT))

model.add(Dense(510))
model.add(Activation('relu'))
model.add(Dropout(DROPOUT))


#model.add(Dense(500))
#model.add(Activation('relu'))
#model.add(Dropout(DROPOUT))

model.add(Dense(1))
model.add(Activation('sigmoid'))
#model.summary()

# Loss Train
model.compile(loss='binary_crossentropy',optimizer=OPTIMIZER,metrics=['accuracy'])

checkpoint = ModelCheckpoint( filepath=os.path.join(MODEL_DIR, "model-{epoch:02d}.h5"),save_best_only=True) 

history = model.fit(X_train, Y_train,batch_size=BATCH_SIZE, epochs=NB_EPOCH,verbose=VERBOSE,
                    validation_split=VALIDATION_SPLIT,callbacks=[checkpoint])
# Test

score = model.evaluate(X_test, Y_test, verbose=VERBOSE)
print("Test score:", score[0])
print('Test accuracy:', score[1])
print('BatchSize,h1,h2:',BATCH_SIZE ,N_HIDDEN ,N_HIDDEN1 )
```
