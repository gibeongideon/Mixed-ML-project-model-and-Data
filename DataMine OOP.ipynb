```python


class VflDataMine:
    def set_url(self):
        raise NotImplementedError()
    
    def get_page(self):
        print('loading page...',end='')
        sleep(2)
        self.driver = webdriver.Chrome(executable_path="/home/gsociety/Desktop/solen/chromedriver")#,chrome_options=options)
        self.driver.get(self.url)
        sleep(3)
        print('Loading complete!',end='')
        self.pageSource=(self.driver.page_source).encode('utf-8')
        
    def get_table(self):
        try:
            self.table = pd.read_html(self.pageSource)
        except:
            self.driver.close()
            print('No table found, retring in  4 Minutes')
            sleep(4)
            self.mine_process_data()
            pass
        
    def process_table(self):
        raise NotImplementedError()
        
    def output_table_as_file(self):
        raise NotImplementedError()
    
    def mine_process_data(self):
        for self.url in self.r_links:
    
            self.get_page()
            print('geting table')
            sleep(3)
            self.get_table()
            print('processing table')
            sleep(3)
            self.process_table()
            print('saving table')
            sleep(3)
            self.output_table_as_file()
        
```

```python
class VflOddsMine(VflDataMine):
    def __init__(self,r_links):
        self.r_links=r_links
        
    def process_table(self):
        try:
            
            fList=[]
            fList2=[]
            for n in range(0,15,2):
                x=[self.table[0].iloc[:, 0][n],self.table[0].iloc[:, 1][n]]
                y=(self.table[0].iloc[:, 0][n+1])
                String1=y.replace(',','.')
                revString1 = re.sub(r'\s\s', ",", String1).split(",")
                Rev=revString1+x
                d,d2=rL(Rev)
                fList.append(d)
                fList2.append(d2)


            dF=pd.DataFrame(fList)
            dF2=pd.DataFrame(fList2)

            fList.clear()
            fList2.clear()

            self.DF=dF.fillna(0)  
            self.DF2=dF2.fillna(0)
        except:
            print('Data not processed!')
            self.driver.close()
            pass
        
    def output_table_as_file(self):
        try:
            
            with open('/home/gsociety/Desktop/The Model/STORAGE/liveDataN.csv', 'a') as f:
                self.DF.to_csv(f, index=False, header=False)

            with open('/home/gsociety/Desktop/The Model/STORAGE/liveDataRAWN.csv', 'a') as f:
                self.DF2.to_csv(f, index=False, header=False)

            print('Data saving DONE!....',end='')
            self.driver.close()
        except:
            print('saving error')
            pass
```
